–ò–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤–∏–¥–Ω–æ, –∫–∞–∫ –ø—Ä–æ–ø—É—Å–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏–ª–∏—Å—å –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–∏–ø–∞ –¥–æ—Ö–æ–¥–∞. –ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ —Ä–∞–±–æ—Ç–Ω–∏–∫–æ–≤ –ø–æ –Ω–∞–π–º—É.

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –¥–æ–ª—é –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö **days_employed** –∏ **total_income**, —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ç–∏–ø –ø—Ä–æ–ø—É—Å–∫–æ–≤.

```python
# –ù–∞–π–¥—ë–º –¥–æ–ª—é –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ days_employed
# –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–∞–Ω–Ω—ã—Ö 'income_type':
for i in df['income_type'].unique():
    data = df.query('income_type == @i')['days_employed']
    missing_share = 1 - data.count() / len(data)
    print(f'–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "—Ç–∏–ø –¥–æ—Ö–æ–¥–∞" "{i}": {missing_share:.1%}')
```

```python
# –ù–∞–π–¥—ë–º –¥–æ–ª—é –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ total_income
# –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–∞–Ω–Ω—ã—Ö 'income_type':
for i in df['income_type'].unique():
    data = df.query('income_type == @i')['total_income']
    missing_share = 1 - data.count() / len(data)
    print(f'–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "—Ç–∏–ø –¥–æ—Ö–æ–¥–∞" "{i}": {missing_share:.1%}')
```

–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∞–Ω–∞–ª–∏–∑–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–æ–ø—É—Å–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏–ª–∏—Å—å —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º **employee, retiree, business, civil servant**

–í—ã—Å–æ–∫–∞—è –¥–æ–ª—è –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ **entrepreneur**, –Ω–æ –∫ —ç—Ç–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç —Ç–æ–ª—å–∫–æ 2 –∫–ª–∏–µ–Ω—Ç–∞ (–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∏–∂–µ).

–ú–æ–∂–µ–º –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö **days_employed –∏ total_income** –∏–º–µ—é—Ç —Å–ª—É—á–∞–π–Ω—ã–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä.

```python
df['income_type'].value_counts()
```

### –ó–∞–º–µ–Ω–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤, –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

–ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–∏–≤–µ–¥–µ–º –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫–µ **education**. –î–∞–Ω–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è –Ω–∞–º –¥–ª—è –∑–∞–º–µ–Ω—ã –ø—Ä–æ–ø—É—Å–∫–æ–≤.

```python
df['education'] = df['education'].str.lower()

# –ò–∑—É—á–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
for column in ['education']:
    print(f'–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ {column}:')
    print(df[column].unique())
```

–ó–∞–ø–æ–ª–Ω–∏–º –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ **days_employed** —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ transform –º–µ–¥–∏–∞–Ω–æ–π –ø–æ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º —Å—Ç–æ–ª–±—Ü–æ–≤ **gender, dob_years, income_type**, —Ç.–∫. –æ–ø—ã—Ç —Ä–∞–±–æ—Ç—ã –º–æ–∂–µ—Ç –∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–ª–æ–º –∫–ª–∏–µ–Ω—Ç–∞, –≤–æ–∑—Ä–∞—Å—Ç–æ–º –∏ —Ç–∏–ø–æ–º –¥–æ—Ö–æ–¥–∞ –∫–ª–∏–µ–Ω—Ç–∞.

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ .transform():
df['days_employed'] = df['days_employed'].fillna(df.groupby(['gender','dob_years','income_type'])['days_employed'].transform('median'))

# –ü—Ä–æ–≤–µ—Ä–∏–º, –æ—Å—Ç–∞–ª–∏—Å—å –ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ:
df.loc[df['days_employed'].isna()].count().sum()
```

–ó–∞–ø–æ–ª–Ω–∏–º –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ **total_income** —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ transform –º–µ–¥–∏–∞–Ω–æ–π –ø–æ –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º —Å—Ç–æ–ª–±—Ü–æ–≤ **education, dob_years, income_type**, —Ç.–∫. –µ–∂–µ–º–µ—Å—è—á–Ω—ã–π –¥–æ—Ö–æ–¥ –º–æ–∂–µ—Ç –∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º, –≤–æ–∑—Ä–∞—Å—Ç–æ–º –∏ —Ç–∏–ø–æ–º –¥–æ—Ö–æ–¥–∞ –∫–ª–∏–µ–Ω—Ç–∞.

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ .transform():
df['total_income'] = df['total_income'].fillna(df.groupby(['education','dob_years','income_type'])['total_income'].transform('median'))

# –ü—Ä–æ–≤–µ—Ä–∏–º, –æ—Å—Ç–∞–ª–∏—Å—å –ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ:
df.loc[df['total_income'].isna()].count().sum()
```

- `gender` ‚Äî –ø–æ–ª –∫–ª–∏–µ–Ω—Ç–∞;
- `income_type` ‚Äî —Ç–∏–ø –¥–æ—Ö–æ–¥–∞ –∫–ª–∏–µ–Ω—Ç–∞.

```python
# –ó–∞–º–µ–Ω—è–µ–º –Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞ –Ω–∞ –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:
df['dob_years'] = df['dob_years'].replace(0, np.nan)

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ .transform() –¥–ª—è –∑–∞–º–µ–Ω—ã –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —É—Å–ª–æ–≤–∏—é:
df['dob_years'] = df['dob_years'].fillna(df.groupby(['gender','income_type'])['dob_years'].transform('median'))

# –ü—Ä–æ–≤–µ—Ä–∏–º –¥–æ–ª—é –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ dob_years:
df['dob_years'].isna().mean()
```

- –û—á–µ–Ω—å –Ω–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –Ω–∏–∂–µ 14 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞.

- –í—ã—Å–æ–∫–∏–π –¥–æ—Ö–æ–¥ ‚Äî –ª—é–¥–∏, –ø–æ–ª—É—á–∞—é—â–∏–µ –±–æ–ª—å—à–µ 78 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—è –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–∞.



### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —É—Ä–æ–≤–Ω—é –¥–æ—Ö–æ–¥–∞


```python
# –ó–∞–¥–∞–¥–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø—è—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–π:
cat_names = ['–ë–µ–∑ –¥–æ—Ö–æ–¥–∞','–û—á–µ–Ω—å –Ω–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥','–ù–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥', '–°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥', '–í—ã—Å–æ–∫–∏–π –¥–æ—Ö–æ–¥']

# –†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ 6 –∫–∞—Ç–µ–≥–æ—Ä–∏–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π:
df['total_income_category'] = pd.qcut(df['total_income'], q=[0, 0.1, 0.14, 0.34, 0.78, 1], labels=cat_names)

# –ò–∑—É—á–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –Ω–æ–≤–æ–º —Å—Ç–æ–ª–±—Ü–µ:
df['total_income_category'].value_counts()
```

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å —Å—Ä–µ–¥–Ω–∏–º –¥–æ—Ö–æ–¥–æ–º, –∞ –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ –∫ –∫–ª–∏–µ–Ω—Ç–æ–≤ - –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –±–µ–∑ –¥–æ—Ö–æ–¥–∞

### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É


1. –†–∞–∑–¥–µ–ª–∏—Ç–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –Ω–∞ –¥–≤–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É: –¥–æ 40 –ª–µ—Ç –∏ –ø–æ—Å–ª–µ. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∫–æ–ª–æ–Ω–∫–µ `age_category`.



```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É:
def age_cat(dob_years):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É:
    - –î–æ 40 –ª–µ—Ç
    - –ü–æ—Å–ª–µ 40 –ª–µ—Ç
    """

    if dob_years < 40:
      return '–¥–æ 40 –ª–µ—Ç'

    elif dob_years >= 40:
      return '40 –ª–µ—Ç –∏ —Å—Ç–∞—Ä—à–µ'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É:
df['age_category'] = df['dob_years'].apply(age_cat)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['age_category'].value_counts()
```

### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π (–±–æ–ª–µ–µ –ø–æ–¥—Ä–æ–±–Ω–∞—è)


1. –†–∞–∑–¥–µ–ª–∏—Ç–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π: –±–µ–∑ –¥–µ—Ç–µ–π, –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö, –æ—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∫–æ–ª–æ–Ω–∫–µ `childrens_category`.



```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
def children_cat(children):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
    - 0 - –±–µ–∑ –¥–µ—Ç–µ–π
    - 1-2 - –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö
    - 3 –∏ –±–æ–ª—å—à–µ - –æ—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ
    """

    if children == 0:
      return '–±–µ–∑ –¥–µ—Ç–µ–π'

    elif children >= 1 and children <= 2:
      return '–æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –¥–≤—É—Ö'

    elif children >= 3:
      return '–æ—Ç —Ç—Ä—ë—Ö –∏ –±–æ–ª—å—à–µ'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
df['children_category'] = df['children'].apply(children_cat)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['children_category'].value_counts()
```

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –±–û–ª—å—à–∞—è —á–∞—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π

### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π (–º–µ–Ω–µ–µ –ø–æ–¥—Ä–æ–±–Ω–∞—è)

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
def children_short_cat(children):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
    - 0 - –±–µ–∑ –¥–µ—Ç–µ–π
    - 1 –∏ –±–æ–ª—å–µ–µ 1-2 - —Å –¥–µ—Ç—å–º–∏
    """

    if children == 0:
      return '–±–µ–∑ –¥–µ—Ç–µ–π'

    elif children >= 1:
      return '—Å –¥–µ—Ç—å–º–∏'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–µ—Ç–µ–π:
df['children_short_category'] = df['children'].apply(children_short_cat)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['children_short_category'].value_counts()
```

### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ü–µ–ª—è–º –∫—Ä–µ–¥–∏—Ç–æ–≤–∞–Ω–∏—è

–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ü–µ–ª—è–º –∫—Ä–µ–¥–∏—Ç–æ–≤–∞–Ω–∏—è

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Ü–µ–ª—è–º –∫—Ä–µ–¥–∏—Ç–æ–≤–∞–Ω–∏—è:
def purpose_cat(purpose):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ü–µ–ª–∏ –∫—Ä–µ–¥–∏—Ç–æ–≤–∞–Ω–∏—è:
    - –ü–æ–∫—É–ø–∫–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏
    - –ü–æ–∫—É–ø–∫–∞ –º–∞—à–∏–Ω—ã
    - –ü–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è
    - –û—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è —Å–≤–∞–¥—å–±—ã
    - –†–µ–º–æ–Ω—Ç –∫–≤–∞—Ä—Ç–∏—Ä—ã
    """
    if purpose in ['building a property','building a real estate','buy commercial real estate','buy real estate','buy residential real estate','buying property for renting out',
                     'construction of own property','property','purchase of my own house','purchase of the house','purchase of the house for my family','real estate transactions',
                     'transactions with commercial real estate','transactions with my real estate']:
      return 'to buy / build a property'

    if purpose in ['buying a second-hand car''buying my own car','car','car purchase','cars','purchase of a car','second-hand car purchase','to buy a car','to own a car']:
      return 'to buy a car'

    elif purpose in ['education','getting an education','getting higher education','profile education','supplementary education',
                     'to become educated','to get a supplementary education','university education','going to university']:
      return 'to get an education'

    elif purpose in ['having a wedding','to have a wedding','wedding ceremony']:
      return 'to have a wedding'

    elif purpose in ['housing','housing renovation','housing transactions']:
      return 'to renovate a house'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ü–µ–ª–∏ –∫—Ä–µ–¥–∏—Ç–æ–≤–∞–Ω–∏—è:
df['purpose_category'] = df['purpose'].apply(purpose_cat)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['purpose_category'].value_counts()
```

–ò–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –≤–∏–¥–Ω–æ, —á—Ç–æ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∫—Ä–µ–¥–∏—Ç–æ–≤ –≤ –Ω–∞—à–µ–º –±–∞–Ω–∫–µ –æ—Ñ–æ—Ä–º–ª–µ—è—Ç—Å—è –¥–ª—è —Ü–µ–ª–µ–π –ø–æ–∫—É–ø–∫–∏ –∏–ª–∏ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏.

### –ê–Ω–∞–ª–∏–∑ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–æ–ª—É –∏ –Ω–∞–ª–∏—á–∏—é –¥–æ–ª–≥–æ–≤

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–æ–ª—É –∏ –Ω–∞–ª–∏—á–∏—é –¥–æ–ª–≥–æ–≤:
def gender_debt_cat(df):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–æ–ª—É –∏ –Ω–∞–ª–∏—á–∏—é –¥–æ–ª–≥–æ–≤:
    - "M, 0" - –º—É–∂—á–∏–Ω—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤
    - "M, 1" - –º—É–∂—á–∏–Ω—ã —Å –¥–æ–ª–≥–∞–º–∏
    - "F, 0" - –∂–µ–Ω—â–∏–Ω—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤
    - "F, 1" - –∂–µ–Ω—â–∏–Ω—ã —Å –¥–æ–ª–≥–∞–º–∏
    """

    gender = df['gender']
    debt = df['debt']

    if gender == 'M' and debt == 0:
      return '–ú—É–∂—á–∏–Ω—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤'

    elif gender == 'M' and debt == 1:
      return '–ú—É–∂—á–∏–Ω—ã —Å –¥–æ–ª–≥–∞–º–∏'

    elif gender == 'F' and debt == 0:
      return '–ñ–µ–Ω—â–∏–Ω—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤'

    elif gender == 'F' and debt == 1:
      return '–ñ–µ–Ω—â–∏–Ω—ã —Å –¥–æ–ª–≥–∞–º–∏'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–æ–ª—É –∏ –Ω–∞–ª–∏—á–∏—é –¥–æ–ª–≥–æ–≤:
df['gender_debt_category'] = df.apply(gender_debt_cat, axis=1)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['gender_debt_category'].value_counts()
```

```python
pivot_education = df.pivot_table(
    index=['education','age_category'],
    columns='debt',
    values='total_income',
    aggfunc='count',
    fill_value=0)

# –ü–µ—Ä–µ–∏–º–µ–Ω—É–µ–º —Å—Ç–æ–ª–±—Ü—ã
pivot_education_modified = pivot_education.rename(columns={
                   0: '–ö–ª–∏–µ–Ω—Ç—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤',
                   1: '–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏'
                   })

# –î–æ–±–∞–≤–∏–º —Ä–∞—Å—á–µ—Ç–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü
pivot_education_modified['–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏, %'] = round(
    pivot_education_modified['–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏'] / (pivot_education_modified['–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏'] + pivot_education_modified['–ö–ª–∏–µ–Ω—Ç—ã –±–µ–∑ –¥–æ–ª–≥–æ–≤']) * 100, 1
    )

# –û—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç:
pivot_education_modified = pivot_education_modified.sort_values(by=['education','–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏, %'])

# –í—ã–≤–µ–¥–µ–º –∏—Ç–æ–≥
pivot_education_modified
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏

```python
# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(10, 6))

# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞
barplot = sns.barplot(
    x='education',
    y='–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏, %',
    data=pivot_education_modified,
    hue='age_category',
    palette='coolwarm',
    errorbar=None
    )

# –ü–æ–≤–æ—Ä–æ—Ç –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–∏ x
plt.xticks(rotation=20)

# –£–∫–∞–∑—ã–≤–∞–µ–º –∏–º—è –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –∏–º–µ–Ω–∞ –æ—Å–µ–π
barplot.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–∏–∫–æ–≤ —Å—Ä–µ–¥–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ —Ä–∞–∑—Ä–µ–∑–µ —É—Ä–æ–≤–Ω—è –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞', fontsize=12)
barplot.set_xlabel('–£—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è', fontsize=14)
barplot.set_ylabel('–ö–ª–∏–µ–Ω—Ç—ã —Å –¥–æ–ª–≥–∞–º–∏, %', fontsize=14)

# –¶–∏–∫–ª –¥–ª—è –≤—ã–≤–æ–¥–∞ –ø–æ–¥–ø–∏—Å–µ–π –∫–æ –≤—Å–µ–º –±–∞—Ä–ø–ª–æ—Ç–∞–º
for i in barplot.containers:
  barplot.bar_label(i,)

# –í—ã–≤–æ–¥–∏–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.show()
```

### –ì–∏–ø–æ—Ç–µ–∑–∞ ‚Ññ2 - –æ–¥–∏–Ω–æ–∫–∏–µ / —Å–µ–º–µ–π–Ω—ã–µ –º—É–∂—á–∏–Ω—ã


***–ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫—É—é –≥–∏–ø–æ—Ç–µ–∑—É***: –æ–¥–∏–Ω–æ–∫–∏–µ –º—É–∂—á–∏–Ω—ã —Å –Ω–∏–∑–∫–∏–º –¥–æ—Ö–æ–¥–æ–º —á–∞—â–µ –æ–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –¥–æ–ª–∂–Ω–∏–∫–∞–º–∏, —á–µ–º —Å–µ–º–µ–π–Ω—ã–µ –º—É–∂—á–∏–Ω—ã —Å–æ —Å—Ä–µ–¥–Ω–∏–º –¥–æ—Ö–æ–¥–æ–º.

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–µ–º–µ–π–Ω–æ–≥–æ —Å—Ç–∞—Ç—É—Å–∞ –∏ id

df_raw[['family_status', 'family_status_id']].value_counts()
```

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º–æ–∂–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º—É–∂—á–∏–Ω —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º, –æ–¥–∏–Ω–æ–∫–∏–µ - id 2,3,4. –°–µ–º–µ–π–Ω—ã–µ - 0 –∏ 1.

```python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é:
def family_status_cat(family_status_id):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é:
    - married, civil partnership - —Å–µ–º–µ–π–Ω—ã–µ
    - unmarried, divorced, widow / widower - –æ–¥–∏–Ω–æ–∫–∏–µ
    """

    if family_status_id >= 0 and family_status_id < 2:
      return '—Å–µ–º–µ–π–Ω—ã–π'

    elif family_status_id >= 2 and family_status_id <= 4:
      return '–æ–¥–∏–Ω–æ–∫–∏–π'

# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é:
df['family_status_category'] = df['family_status_id'].apply(family_status_cat)

# –í—ã–≤–µ–¥–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —ç—Ç–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:
df['family_status_category'].value_counts()
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–µ–π –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–æ–ª—É, —Å–µ–º–µ–π–Ω–æ–º—É —Å—Ç–∞—Ç—É—Å—É –∏ —É—Ä–æ–≤–Ω—é –¥–æ—Ö–æ–¥–∞
